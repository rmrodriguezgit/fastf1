```python
import requests
import pandas as pd
import json
from pandas import json_normalize
```

```python
year = input('Temporada : ')
round = input('Round :')
#driver = input('Driver :')
driver= 'verstappen'
```

```python
url = "http://ergast.com/api/f1/"
query = "results.json"
urlquery = url + str(year) + '/'+str(round)+'/'+query
urlquery
```

```python
response = requests.get(urlquery)
data = response.json()
print(json.dumps(data, indent=4))
```

```python
drivers = data['MRData']['RaceTable']['Races'][0]['Results']
print(json.dumps(drivers, indent=4))

```

```python
df = pd.DataFrame(drivers)
df
```

```python
df2 = pd.json_normalize(drivers)
df2
```

```python
results = print(df2[['position','Driver.code','points']])
results
```
