```python
import requests
import pandas as pd
import json
from pandas import json_normalize
```

```python
year = input('Temporada : ')
```

```python
url = "http://ergast.com/api/f1/"
query = "driverStandings.json"
urlquery = url + str(year) + '/'+query
urlquery
```

```python
response = requests.get(urlquery)
data = response.json()
drivers = data['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
drivers
```

```python
df = pd.DataFrame(drivers)
df
```

```python
df2 = pd.json_normalize(drivers)
df2
```

```python
df2.drop(['Constructors'],axis=1)
```

```python
df2.dtypes
```

```python
df2['position'] = df2['position'].astype(int)
df2['points'] = df2['points'].astype(float)
df2['wins'] = df2['wins'].astype(int)
```
