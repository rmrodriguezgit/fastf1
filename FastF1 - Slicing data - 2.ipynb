```python
import fastf1
```

```python
fastf1.Cache.enable_cache("cache")
```

```python
year = int(input('Año: '))
wknd = int(input('Round: '))
print('------ Session Type ------')
print('R - Race')
print('Q - Qualifying')
print('SQ - Sprint Qualifying')
print('FP1 - Practice 1')
print('FP2 - Practice 2')
print('FP3 - Practice 3')
print('--------------------------')
ses = input('Session : ')
driver = input('Piloto : ')
```

```python
session = fastf1.get_session(year, wknd, ses)
session.load()
evento = session.event
fastest_lap = session.laps.pick_fastest()
fastest_lap_driver = session.laps.pick_driver(driver).pick_fastest()
fastest_laps = session.laps.pick_quicklaps()
fastest_laps_driver = session.laps.pick_driver(driver).pick_quicklaps()
tel = fastest_lap.telemetry
tel_driver = fastest_lap_driver.telemetry
```

```python
print('---------- Event ----------------')
print(evento)
print('--------- End Event -------------')
```

```python
print('---------------- Fastest Lap -------------------')
print(fastest_lap)
print('-------------- End Fastest Lap -----------------')
```

```python
print(f'---------------- Fastest Lap {driver} -------------------')
print(fastest_lap_driver)
print(f'-------------- End Fastest Lap {driver} -----------------')
```

```python
print(f'---------------- Fastest Lap {driver} -------------------')
print(fastest_laps_driver)
print(f'-------------- End Fastest Lap {driver} -----------------')
```

```python
print('---------------------------- Fastest Laps --------------------------------')
print(fastest_laps)
print('-------------------------- End Fastest Laps ------------------------------')
```

```python
print('--------------------------- Telemetry -------------------------------')
print(tel)
print('------------------------- End Telemetry -----------------------------')
```

```python
print(f'--------------------------- Telemetry {driver} --------------------------')
print(tel_driver)
print(f'------------------------- End Telemetry {driver} ------------------------')
```
